# What is Matplotlib?
Matplotlib is a Python data visualization library used to create charts and graphs such as:
* Line plots
* Bar charts
* Histograms
* Pie charts
* Scatter plots
* Box plots

## Types of Data:
1. Numerical Data
2. Categorical Data

```python
# import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('default')
```

```python
# import data sets
batsman = pd.read_csv('/content/sharma-kohli.csv')
batter = pd.read_csv('/content/batter.csv')
```

# 2D Line Plot
A 2D line plot is a graph that shows the relationship between two variables:
* One variable on the X-axis (horizontal)
* One variable on the Y-axis (vertical)

The data points are connected with a line.

It is called 2D because it has only:
* X dimension
* Y dimension

## When Do We Use It?
A 2D line plot is mostly used to show:
* üìä Trends over time
* üìà Growth or decline
* üîÅ Continuous data
* üìâ Comparisons


```python
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]

plt.plot(year,price)
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

# Customization

```python
# plotting using multiple plots
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='solid',linewidth=3,marker='D',label='rohit')
plt.plot(batsman['index'],batsman['V Kohli'],color='blue',linestyle='dashed',linewidth=2,marker='o',markersize=10,label='virat')

plt.title('Career Comparison')
plt.xlabel('Seasons')
plt.ylabel('Runs Scorred')
plt.legend()
```

```python
# grid
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='solid',linewidth=3,marker='D',label='rohit')
plt.plot(batsman['index'],batsman['V Kohli'],color='blue',linestyle='dashed',linewidth=2,marker='o',markersize=10,label='virat')

plt.title('Career Comparison')
plt.xlabel('Seasons')
plt.ylabel('Runs Scorred')
plt.legend()

plt.grid()

# plt.show() -> use to show graphs in console.
```

```python
# limiting axes
# outlier
price = [48000,54000,57000,49000,47000,4500000]
year = [2015,2016,2017,2018,2019,2020]
plt.plot(year,price)
plt.ylim(0,70000)
```

# Scatter Plot
A scatter plot is a graph used to show the relationship between two numerical variables.

* One variable is placed on the X-axis
* The other variable is placed on the Y-axis
* Each point represents one observation

Unlike a line plot, the points are not connected.

## When Do We Use It?
* üìà Identify correlation (positive or negative)
* üîç Detect outliers
* üìä Understand relationships between variables
* ü§ñ Prepare for regression analysis

```python
x = np.linspace(-10,10,50)
y = 10*x + 3 + np.random.randint(0,300,50)
```

```python
plt.scatter(x,y)
```

```python
df = batter.head(50)
plt.scatter(df['avg'],df['strike_rate'], color='green')

plt.xlabel('Average')
plt.ylabel('Strike Rate')

plt.title('Avg and SR Analysis of top 50 Batsmans')
```

```python
tips = sns.load_dataset('tips')
tips.head(3)
```

```python
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*10)

plt.xlabel('Total Bill')
plt.ylabel('Tip')
```

```python
# scatter plot usint plt.plot
# Faster method than scatter
plt.plot(tips['total_bill'],tips['tip'],'o')
```

# Bar Chart
A bar chart (or bar graph) is a graph used to compare different categories using rectangular bars.
* Categories are shown on the X-axis
* Values are shown on the Y-axis
* The height (or length) of each bar represents the value

## Why Bar Charts Are Important in Data Analytics
Bar charts help to:
* üìä Compare categories easily
* üìà Show differences clearly
* üßæ Present business reports
* üíº Analyze performance (sales, profit, users, etc.)

## Types of Bar Chart
* Vertical Bar Chart (most common)
* Horizontal Bar Chart ‚Üí plt.barh()
* Grouped Bar Chart
* Stacked Bar Chart

```python
# vertical bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children)
```

```python
# horizontal bar chart
plt.barh(colors,children)
```
