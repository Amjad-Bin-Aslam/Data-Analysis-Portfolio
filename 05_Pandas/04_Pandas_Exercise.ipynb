# concat()
pd.concat() is a pandas function used to combine two or more DataFrames (or Series) by stacking them either vertically (row-wise) or horizontally (column-wise).

* Papers on top of each other → row-wise
* Papers side by side → column-wise

### Parameters:

* objs → list of DataFrames
* axis=0 → stack rows (default)
* axis=1 → stack columns
* ignore_index=True → reset index

### When to use:
* Adding more rows to dataset
* Combining monthly data files
* Merging datasets with same structure
* Stacking features horizontally

```python
regs = pd.concat([nov,dec],ignore_index=True)
regs
```

```python
# Multi-index DataFrame
multi = pd.concat([nov,dec],keys=['Nov','Dec'])
# get the first_element from nov dataFrame
multi.loc[('Nov',0)]
```

```python
# concat() -> side-by-side , column-wise(Horizantly)
pd.concat([nov,dec],axis=1).shape
```

```python
df1 = pd.DataFrame({'A': [1,2], 'B': [3,4]})
df2 = pd.DataFrame({'C':[5,6,7]})
pd.concat([df1,df2],axis=1)
```

# merge()
pd.merge() is a pandas function used to combine two DataFrames based on a common column or key, similar to SQL JOIN operations.

### Types:
1. inner
2. left
3. right
4. outer

### Parameters:
* left → first DataFrame
* right → second DataFrame
* on → common column name
* how → type of join

### When to use:
* Combining related tables
* Joining datasets with a key column
* Working like SQL joins
* Merging relational data
